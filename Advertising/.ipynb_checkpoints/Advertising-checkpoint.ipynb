# Hypothesis Testing

```python
# State the null and alternative hypothesis
# H0: The coefficient of TV advertising is not significantly different from 0
# Ha: The coefficient of TV advertising is significantly different from 0

# Set the significance level (alpha value)
alpha = 0.05

# Create the X and Y matrices for the linear regression model
X = df["TV"]
Y = df["Sales"]

# Add a constant to the X matrix for the intercept term
X = sm.add_constant(X)

# Fit the linear regression model
model = sm.OLS(Y, X).fit()

# Get the p-value for the TV advertising coefficient
p_value = model.pvalues["TV"]

# Print the p-value
print("p-value:", p_value)

# Compare the p-value to the significance level and make a decision
if p_value < alpha:
    print("Reject the null hypothesis. The coefficient of TV advertising is significantly different from 0.")
else:
    print("Fail to reject the null hypothesis. The coefficient of TV advertising is not significantly different from 0.")
```
