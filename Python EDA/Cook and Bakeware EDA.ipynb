<b>Splitting strings in the cookwarebakewares_name Column<b>

```python
ware_column_names = dict()
for ware in data['cookwarebakewares_name']:
    if type(ware) == str:
        for item in ware.split(","):
            if ware_column_names.get(item) is None:
                ware_column_names[item] = 1
            else:
                ware_column_names[item] += 1
print(ware_column_names)   
```
