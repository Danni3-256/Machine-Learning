```python
advertising_dict = {'TV':{1:230.1,2:44.5,3:17.2,4:151.5,5:180.8},'Radio':{1:37.8,2:39.3,3:45.9,4:41.3,5:10.8},'Newspaper':{1:69.2,2:45.1,3:69.3,4:58.5,5:58.4},'Sales':{1:22.1,2:10.4,3:9.3,4:18.5,5:12.9}}
advertising_data_df = pd.DataFrame(advertising_dict)
advertising_data_df
```

```python
advertising_data_df.size
```

```python
advertising_data_df.columns
```

```python
#Feature object
X_feature = advertising_data_df[['TV', 'Radio', 'Newspaper']]
X_feature
```

```python
Y_target = advertising_data_df[['Sales']]
Y_target
```

# Splitting the dataset

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X_feature,Y_target,random_state=1)
```

```python
print (x_train.shape)
print (y_train.shape)
print (x_test.shape)
print (y_test.shape)
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
linreg = LinearRegression()
linreg.fit(x_train,y_train)
```

# Intercept and Coefficients
print(linreg.intercept_)
print(linreg.coef_)

```python
y_pred = linreg.predict(x_test)
y_pred
```

# Test model accuracy using MSE

```python
from sklearn import metrics
```

```python
print(np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
```

# Web Scraping With Beautiful Soup

Web Scraping is a computer Software technique of extracting information from websites in an automated fashion

</br>
<b>Parser</b>
</br>
Parsing is a technique used to read and extract data from the web
</br>
A parser interpretes information from a web document

<b>Beautifu Soup</b>
</br>
Beautiful soup is an efficient tool for dissecting documents and extracting information from the web and converts all incoming documents to Unicode automatically.
</br>
It also transforms a complex HTML document into a complex tree of python objects

```python
from bs4 import BeautifulSoup
import requests
```

```python
#url for web scraping
url = 'http://www.simplilearn.com'
```

```python
result = requests.get(url)
webpage = result.content
sl_soup = BeautifulSoup(webpage,'html.parser')
```

```python
#Close result object
result.close()
```

```python
#View content
sl_soup.contents
```

```python
#Well formatted output
print(sl_soup.prettify())
```

```python
#View Head
sl_soup.head
```

```python
sl_soup.title
```

```python
#All Links
for href in sl_soup.findAll('a', href = True):
    print(href['href'])
```
